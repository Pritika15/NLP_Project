```python
from textblob import TextBlob
```

```python
my_sentence = TextBlob("I am reading a blog post on AnalyticsVidhya. I am loving it!")
```

```python
#POS TAGGING
#POS tagging is the task of labeling each word in a phrase with its proper part of speech.
print(my_sentence.tags)
```

```python
#Sentiment Analysis can assist us in determining the mood and feelings 
#The polarity and subjectivity of a statement are returned by TextBlob. The range of polarity is [-1,1], with -1 indicating a negative sentiment and 1 indicating a positive sentiment. 
#Emoticons, exclamation marks, and emojis, for example. subjectivity falls under the numeric range of [0,1].
#The degree of personal opinion and factual information in a text is measured by subjectivity. Because of the text‚Äôs heightened subjectivity, it contains personal opinion rather than factual information. There‚Äôs one more setting in TextBlob: intensity. The ‚Äòintensity‚Äô is used by TextBlob to calculate subjectivity. The intensity of a word influences whether it modifies the next word. Adverbs are used as modifiers in English.
#The polarity score ranges from -1.0 to 1.0 and the subjectivity ranges from 0.0 to 1.0 where 0.0 is an objective statement and 1 is a subjective statement.
my_sentence.sentiment
```

```python
df=pd.read_csv('679071.csv')
df
```

```python
Polarity=[]
subjectivity=[]
for textdata in df["LemmatizeText"]:
  sent=TextBlob(str(textdata))
  Polarity.append(sent.sentiment[0])
  subjectivity.append(sent.sentiment[1])
print("Polarity:")  
Polarity

```

```python
print("subjectivity:")  
subjectivity   
```

```python
df['rawPolarity'] =Polarity
df['rawSubjectivity']=subjectivity  
```

```python
df.head()
```

```python
#Labelling Tweets into Positive(1) and Negative(0) 
NewPolarity=[]
PositiveTweets=0
NegativeTweets=0
for PolarityVal in df['rawPolarity']:
  if PolarityVal >= 0:
    NewPolarity.append(1)
    PositiveTweets=PositiveTweets+1

  if PolarityVal <0:
    NewPolarity.append(0)
    NegativeTweets=NegativeTweets+1
NewPolarity    

print(PositiveTweets)
print(NegativeTweets)
    
```

```python
df['NewPolarity']= NewPolarity  
df.head()
```
